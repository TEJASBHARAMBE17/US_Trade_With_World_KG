```python
# Import libaries
import pandas as pd
import numpy as np
import requests
import time
from lxml import etree
from bs4 import BeautifulSoup
import json
import re
```

```python
# Get data from url
url = 'https://www.worldometers.info/gdp/gdp-by-country/'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
tree = etree.HTML(str(soup))
tree
```

```python
tree.xpath("//thead/tr/th/text()")
```

```python
tree.xpath("//tbody/tr/td/text()")
```

```python
data_list = tree.xpath("//tbody/tr/td/text()")
data = []
df = pd.DataFrame(tree.xpath("//tbody/tr/td/a/text()"))


for i in range(0, int(len(data_list)/7), 1):
    data.append(data_list[i*7:(i*7)+7])
df = pd.concat([df, pd.DataFrame(data)], axis=1)
df.columns = ['country', 'rank', 'gdp_nominal', 'gdp_abbr', 'gdp_growth', 'population', 'gdp_per_capita', 'global_gdp_share']
df
```

```python
df.to_csv('../data/gdp_population.csv', index=False)
```
