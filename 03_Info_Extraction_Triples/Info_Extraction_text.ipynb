```python
import spacy
import en_core_web_sm, en_core_web_md
import csv, json, re
from spacy import displacy
from spacy.matcher import Matcher
from spacy.tokens import Token


def extract_ECI(bio):
    nlp = spacy.load("en_core_web_md")
    nlp.add_pipe("merge_entities", after="ner")
    doc = nlp(bio)

    pattern_0_0 = [

        {"ENT_TYPE": "GPE"},
        {"OP": "{,2}"},
        {"LEMMA": "rank"},
        {"ENT_TYPE": "CARDINAL"},
        {"OP": "{,4}"},
        {"TEXT": {"IN": ["Economic"]}}
    ]
  
    patterns = [
        pattern_0_0,
    ]
    matcher = Matcher(nlp.vocab)
    matcher.add("birthplace_syntactical", patterns)

    matches = matcher(doc)  # multiple matches
    span=[]
    if matches:
        for i in matches[:1]:
            span.append(doc[i[1] : i[2]])
        
        return span
    else:
        return None

def extract_ECI_value(bio):
    nlp = spacy.load("en_core_web_md")
    nlp.add_pipe("merge_entities", after="ner")
    doc = nlp(bio)

    pattern_0_0 = [

        {"lower": "("},
        {"lower": "eci"},
        {"OP": "{,2}"},
        {"lower": ")"},
            ]
  
    patterns = [
        pattern_0_0,
    ]
    matcher = Matcher(nlp.vocab)
    matcher.add("birthplace_syntactical", patterns)

    matches = matcher(doc)  # multiple matches
    span=[]
    if matches:
        for i in matches[:1]:
            span.append(doc[i[1] : i[2]])
            
        return span
    else:
        return None

def extract_Exports_rank(bio):
    nlp = spacy.load("en_core_web_md")
    nlp.add_pipe("merge_entities", after="ner")
    doc = nlp(bio)

    pattern_0_0 = [
        {"ENT_TYPE": "CARDINAL"},
        {"OP": "{,4}"},
        {"TEXT": {"IN": ["exports"]}}
    ]
  
    patterns = [
        pattern_0_0,
    ]
    matcher = Matcher(nlp.vocab)
    matcher.add("birthplace_syntactical", patterns)

    matches = matcher(doc)  # multiple matches
    span=[]
    if matches:
        for i in matches[:1]:
            span.append(doc[i[1] : i[2]])
        return span
    else:
        return None
```

```python
output1=[]
output2=[]
output3=[]

for detail in df['exports_details']:
    output1.append(extract_ECI(detail))
    output2.append(extract_ECI_value(detail))
    output3.append(extract_Exports_rank(detail))
output1,output2,output3
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp('Afghanistan-United States  In 2020,  Afghanistan  exported $15M to  United States . The main products that  Afghanistan  exported to  United States  are  Knotted Carpets  ($4.81M),  Grapes  ($1.31M),  and  Dried Fruits  ($996k). During the last 25 years the exports of Afghanistan to United States have increased at an annualized rate of 4.09%, from $5.49M in 1995 to $15M in 2020.In 2020,  Afghanistan  did not export any services to  United States .United States-Afghanistan  In 2020,  United States  exported $603M to  Afghanistan . The main products that  United States  exported to  Afghanistan  were  Special purpose motor vehicles  ($101M),  Planes, Helicopters, and/or Spacecraft  ($68M),  and  Aircraft Parts  ($52.2M). During the last 25 years the exports of United States to Afghanistan have increased at an annualized rate of 22.7%, from $3.62M in 1995 to $603M in 2020.In 2020,  United States  did not export any services to  Afghanistan .Comparison  In 2020, Â\xa0 Afghanistan  ranked 109 in the Economic Complexity Index (ECI -1.17), and 140 in total exports ($1.76B). That same year,  United States  ranked 9 in the Economic Complexity Index (ECI 1.56), and 2 in total exports ($1.34T).')

df_spacy=pd.DataFrame()
text__=[]
lemma__=[]
pos__=[]
tag__=[]
dep__=[]
shape__=[]
is_alpha__=[]
is_stop__=[]
is_digit__=[]

for token in doc:
    text__.append(token.text)
    lemma__.append(token.lemma_)
    pos__.append(token.pos_)
    tag__.append(token.tag_)
    dep__.append(token.dep_)
    shape__.append(token.shape_)
    is_alpha__.append(token.is_alpha)
    is_stop__.append(token.is_stop)
    is_digit__.append(token.is_digit)

df_spacy['text_']=text__
df_spacy['lemma_']=lemma__
df_spacy['pos_']=pos__
df_spacy['tag_']=tag__
df_spacy['dep_']=dep__
df_spacy['is_alpha_']=is_alpha__
df_spacy['is_stop_']=is_stop__
df_spacy['is_digit_']=is_digit__

df_spacy.to_csv('extracted_data.csv',index=False)
```
